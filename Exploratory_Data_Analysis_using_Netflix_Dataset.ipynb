**Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.**

```python
netflix_df.columns
```

```python
netflix_df['Release_Date']
```

```python
import datetime
```

```python
netflix_df['Release_Year']=pd.to_datetime(netflix_df['Release_Date']).dt.year
```

```python
netflix_df['Release_Year'].value_counts()
```

```python
netflix_df.groupby('Release_Year').agg({'Title':'count'}).sort_values(by='Title',ascending=False)
```

```python
netflix_df.groupby('Release_Year').agg({'Title':'count'}).sort_values(by='Title',ascending=False)
```

```python
netflix_df.groupby('Release_Year').agg({'Title':'count'}).sort_values(by='Title',ascending=False).plot(kind='bar')
```

```python
import matplotlib.pyplot as plt

year_counts = netflix_df.groupby('Release_Year')['Title'].count().sort_values(ascending=False)
plt.figure(figsize=(12, 6))  # Adjust the size of the figure as per your preference
year_counts.plot(kind='bar', color='steelblue')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Number of Movies per Release Year')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
